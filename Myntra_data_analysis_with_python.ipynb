```python

orders["Month"] = orders["Date"].dt.strftime("%B")
orders
```

```python

orders["Total Price"] = orders["Original Price"] - (orders["Original Price"] * orders["Discount%"])
orders
```

```python

py = orders.groupby("Month").agg({"Total Price":"sum"})
py
```

```python
sns.barplot(x = py.index, y = py["Total Price"], order = ["January", "February", "March",'April',"May","June","July","August","September","October","November", "December"],hue = py.index, palette = "GnBu")
plt.xticks(rotation = 45)
plt.show()
```

```python

df = pd.merge(left = orders, right= products, on = "Product ID", how = "inner")
df
```

```python

py1 = df.groupby("Category").agg({"Total Price":"sum"})
py1
```

```python

plt.pie(py1["Total Price"], labels = py1.index, autopct = "%1.2f%%")
plt.show()
```
